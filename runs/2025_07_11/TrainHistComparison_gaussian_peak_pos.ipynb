```python
import numpy as np
import os, sys
import matplotlib.pyplot as plt
import h5py as h5
import pickle
```

```python
with_dist_grad_path = "losses_hist/NAHEA_CNN_GaussPeak_dist_grad_true_train_hist.pickle"
without_dist_grad_path = "losses_hist/NAHEA_CNN_dist_grad_false_train_hist.pickle"
classical_path = "losses_hist/classical_CNN_kernel_size_5_stride_4_gaussian_peak_train_hist.pickle"
classical_path_learned_channel_collapse = "losses_hist/classical_CNN_kernel_size_5_stride_4_gaussian_peak_leanred_ch_collapse_train_hist.pickle"

with open(with_dist_grad_path, "rb") as f:
    loss_hist_with_dist_grad = pickle.load(f)["loss_hist"]
with open(without_dist_grad_path, "rb") as f:
    loss_hist_without_dist_grad = pickle.load(f)["loss_hist"]
with open(classical_path, "rb") as f:
    loss_hist_classical = pickle.load(f)["loss_hist"]
with open(classical_path_learned_channel_collapse, "rb") as f:
    loss_hist_classical_learned_channel_collapse = pickle.load(f)["loss_hist"]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(np.sqrt(loss_hist_with_dist_grad), label="QNN with learned positions", color='blue')
plt.plot(np.sqrt(loss_hist_without_dist_grad), label="QNN with fixed positions", color='orange')
# plt.plot(np.sqrt(loss_hist_classical), label="Classical CNN (mean channel collapse)", color='green')
plt.plot(np.sqrt(loss_hist_classical_learned_channel_collapse ), label="Classical CNN", color='red')
plt.title("Training loss comparison (Gaussian Peak Pos Finder)")
plt.xlabel("Batch")
plt.ylabel("Loss (RMSE)")
plt.yscale('log')  # Use logarithmic scale for better visibility
plt.legend()
plt.grid()
```
