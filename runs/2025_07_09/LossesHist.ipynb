```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
NAHEA_CNN3_hist_path = "losses_hist/NAHEA_CNN3_train_hist.pickle"
NAHEA_CNN2_hist_path = "losses_hist/NAHEA_CNN2_train_hist.pickle"
classical_CNN_hist_path = "losses_hist/classical_CNN_train_hist.pickle"

def load_hist(path):
    with open(path, "rb") as f:
        return pickle.load(f)
    
NAHEA_CNN2_hist = load_hist(NAHEA_CNN2_hist_path)
NAHEA_CNN3_hist = load_hist(NAHEA_CNN3_hist_path)
classical_CNN_hist = load_hist(classical_CNN_hist_path)
```

```python
fig, ax = plt.subplots(1, figsize=(15, 5))
ax.plot(NAHEA_CNN2_hist["loss_hist"], label="NAHEA CNN2")
ax.plot(NAHEA_CNN3_hist["loss_hist"], label="NAHEA CNN3")
ax.plot(classical_CNN_hist["loss_hist"][:len(NAHEA_CNN2_hist["loss_hist"])], label="Classical CNN")
ax.set_xlabel("Training Step")
ax.set_ylabel("Loss (MSE)")
ax.set_title("Training Losses")
ax.set_yscale("log")
ax.grid(True)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(15, 5))
ax.plot(NAHEA_CNN2_hist["loss_hist"], label="NAHEA CNN2")
ax.plot(NAHEA_CNN3_hist["loss_hist"], label="NAHEA CNN3")
ax.plot(classical_CNN_hist["loss_hist"], label="Classical CNN")
ax.set_xlabel("Training Step")
ax.set_ylabel("Loss (MSE)")
ax.set_title("Training Losses")
ax.set_yscale("log")
ax.grid(True)
ax.legend()
plt.show()
```
